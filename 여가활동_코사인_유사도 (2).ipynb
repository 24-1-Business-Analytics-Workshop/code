```python
import pandas as pd
import numpy as np

file_path = '/content/drive/MyDrive/2024-1/2022_총괄_20240531_30923.csv'
df = pd.read_csv(file_path, encoding='cp949')
df
```

```python
df1 = df.iloc[:, 0:87]
df1
```

```python
columns_list = df1.columns.tolist()
new_columns_list = [str(i+1) for i in range(len(columns_list))]
df1.columns = new_columns_list
df1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df1_array = df1.values

cosine_sim = cosine_similarity(df1_array)

cosine_sim_df = pd.DataFrame(cosine_sim, index=df.index, columns=df.index)
print(cosine_sim_df)
```

```python
for i in range(len(cosine_sim_df)):
    cosine_sim_df.iloc[i, i] = np.nan

print(cosine_sim_df)
```

```python
# 각 행에서 가장 높은 값을 갖는 값과 그에 해당하는 컬럼명을 추출
max_values = cosine_sim_df.max(axis=1)
max_columns = cosine_sim_df.idxmax(axis=1)

# 결과 출력
for i in range(len(cosine_sim_df)):
    print(f"Row {i}: Max Value = {max_values[i]}, Column = {max_columns[i]}")
```

```python
result_df = pd.DataFrame({
    'Max Value': max_values,
    'Column Name': max_columns
})

result_df
```

```python
csv_file_path = 'result_df.csv'
result_df.to_csv(csv_file_path, index=False, encoding='euc-kr')

from google.colab import files
files.download(csv_file_path)
```
