```python
import pandas as pd

# 만족도 6+7+8
file_path = '/content/drive/MyDrive/Colab Notebooks/24-1 Business Analytics Workshop/data/final/2022_총괄_20240524_34131.csv'
df = pd.read_csv(file_path, encoding='cp949')
df
```

```python
# NaN 값 0으로 대체 : 가장 만족스런 여가활동 행에만 nan 존재
df = df.fillna(0)
df
```

```python
df.info()
```

```python
# 변수 타입 전체를 범주형으로 변경
df = df.astype('category')
df.dtypes
```

```python
# 가구 소득 범주 줄이기
df_fin = df.replace({'가구 소득' : 2}, 1)
df_fin = df_fin.replace({'가구 소득' : 3}, 1)
df_fin = df_fin.replace({'가구 소득' : 4}, 2)
df_fin = df_fin.replace({'가구 소득' : 5}, 2)
df_fin = df_fin.replace({'가구 소득' : 6}, 3)
df_fin = df_fin.replace({'가구 소득' : 7}, 3)
df_fin
```

```python
# 클러스터링용 데이터프레임
df_cluster = df_fin.drop(columns=df_fin.columns[0:48])
df_cluster
```

# 군집 분석

```python
pip install kmodes
```

```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
```

```python
# 최적의 k 찾기
cost = []
K = range(1, 15)
for num in K:
    kmode = KModes(n_clusters = num, n_init = 4)
    kmode.fit_predict(df_cluster)
    cost.append(kmode.cost_)

plt.plot(K, cost, 'bx-')
plt.xlabel('number of clusters')
plt.ylabel('cost')
plt.title('elbow method for finding optimal K')
plt.show()
```

k = 5와 k = 7에서 확 꺾이는 것 확인

## 최적의 k = 5일 경우

```python
# 최적의 k = 5일 경우
kmode = KModes(n_clusters = 5, n_init = 4)
clusters_5 = kmode.fit_predict(df_cluster)

df_cluster_5 = df_cluster.copy()
df_cluster_5['clusters_5'] = clusters_5
df_cluster_5
```

```python
# prompt: df_1 DataFrame 사용: 열 요약하지 않고 다 출력되는 코드
pd.set_option('display.max_columns', None)

df_result_5 = df_cluster_5.groupby('clusters_5').describe()
df_result_5
```

```python
df_a_5 = df_cluster_5.loc[df_cluster_5['성별'] == 1, ['clusters_5','성별']]
df_b_5 = df_cluster_5.loc[df_cluster_5['성별'] == 2, ['clusters_5','성별']]

df_c_5 = df_cluster_5.loc[df_cluster_5['연령'] == 1, ['clusters_5','연령']]
df_d_5 = df_cluster_5.loc[df_cluster_5['연령'] == 2, ['clusters_5','연령']]
df_e_5 = df_cluster_5.loc[df_cluster_5['연령'] == 3, ['clusters_5','연령']]
df_f_5 = df_cluster_5.loc[df_cluster_5['연령'] == 4, ['clusters_5','연령']]
df_g_5 = df_cluster_5.loc[df_cluster_5['연령'] == 5, ['clusters_5','연령']]
df_h_5 = df_cluster_5.loc[df_cluster_5['연령'] == 6, ['clusters_5','연령']]
df_i_5 = df_cluster_5.loc[df_cluster_5['연령'] == 7, ['clusters_5','연령']]

df_j_5 = df_cluster_5.loc[df_cluster_5['혼인상태'] == 1, ['clusters_5','혼인상태']]
df_k_5 = df_cluster_5.loc[df_cluster_5['혼인상태'] == 2, ['clusters_5','혼인상태']]
df_l_5 = df_cluster_5.loc[df_cluster_5['혼인상태'] == 3, ['clusters_5','혼인상태']]

df_m_5 = df_cluster_5.loc[df_cluster_5['가구 소득'] == 1, ['clusters_5','가구 소득']]
df_n_5 = df_cluster_5.loc[df_cluster_5['가구 소득'] == 2, ['clusters_5','가구 소득']]
df_o_5 = df_cluster_5.loc[df_cluster_5['가구 소득'] == 3, ['clusters_5','가구 소득']]
```

```python
import matplotlib.pyplot as plt

dfs = [df_a_5, df_b_5]
titles = ['df_a_5', 'df_b_5']

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

for i, df in enumerate(dfs):
    df.groupby('clusters_5')['성별'].count().plot(kind='bar', ax=axes[i])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Clusters')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
dfs = [df_j_5, df_k_5, df_l_5]
titles = ['df_j_5', 'df_k_5', 'df_l_5']

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for i, df in enumerate(dfs):
    df.groupby('clusters_5')['혼인상태'].count().plot(kind='bar', ax=axes[i])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Clusters')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
dfs = [df_m_5, df_n_5, df_o_5]
titles = ['df_m_5', 'df_n_5', 'df_o_5']

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for i, df in enumerate(dfs):
    df.groupby('clusters_5')['가구 소득'].count().plot(kind='bar', ax=axes[i])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Clusters')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
