```python
import matplotlib.pyplot as plt

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

labels = ['남성', '여성', '15~19세', '20대', '30대',	'40대',	'50대',	'60대',	'70세이상']
x = [1,2,3,4,5,6,7,8,9]  # 각 레이블에 해당하는 x 좌표
y_ratios1 = [49.6,	45.8,	53.4,	37.3,	45.4,	50,	48.9,	57.7,	53.9]
y_ratios2 = [46.3,	42.8,	62.6,	41.2,	50.8,	49.1,	34.7,	30.3,	50.1]
y_ratios3 = [48.6, 54.6,	43.8,	59.4,	60.3,	37.3,	53.1,	75.5,	56.6]


plt.scatter(x, y_ratios1, color = 'red', label= '뮤지컬')
plt.scatter(x, y_ratios2, color = 'blue', label= '미술 전시회')
plt.scatter(x, y_ratios3, color = 'green', label= '서양음악')

plt.title('관람 만족도')
plt.xlabel('성별, 나이')
plt.ylabel('비율 (%)')

plt.xticks(x, labels)

# 범례 표시
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
